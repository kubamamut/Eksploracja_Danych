# **Pandas i wykresy**



```python
import pandas as pd

data = {
    "state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
    "year": [2000, 2001, 2002, 2001, 2002, 2003],
    "pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
    }

# Tworzenie DataFrame
frame = pd.DataFrame(data)
```

```python
# Filtracja danych
ohio_data = frame[frame['state'] == 'Ohio']
nevada_data = frame[frame['state'] == 'Nevada']

# Tworzenie wykresu
plt.plot(ohio_data['year'], ohio_data['pop'], marker='o', linestyle='-', color='b',
        label="Ohio")
plt.plot(nevada_data['year'], nevada_data['pop'], marker='o', linestyle='-', color='r',
        label="Nevada")
plt.title('Rok vs Populacja dla Ohio i Nevady')
plt.xlabel('Rok')
plt.ylabel('Populacja [mln]')

# Ustawienie etykiet osi X, aby wywietlay tylko konkretne lata
plt.xticks([2000, 2001, 2002, 2003])

plt.legend()
```

```python
ohio_data.plot(x='year', y='pop', marker='o', linestyle='-', color='b', label="Ohio")
nevada_data.plot(x='year', y='pop', marker='o', linestyle='-', color='r', label="Nevada")
plt.title('Rok vs Populacja dla Ohio i Nevady')
plt.xlabel('Rok')
plt.ylabel('Populacja [mln]')

# Ustawienie etykiet osi X, aby wywietlay tylko konkretne lata
plt.xticks([2000, 2001, 2002, 2003])

plt.legend()
```

```python
# Tutaj wykres w sposb obiektowy

fig, ax = plt.subplots()
ohio_data.plot(x='year', y='pop', marker='o', linestyle='-', color='b', label="Ohio", ax=ax)
nevada_data.plot(x='year', y='pop', marker='o', linestyle='-', color='r', label="Nevada", ax=ax)
ax.set_title('Rok vs Populacja dla Ohio i Nevady')
ax.set_xlabel('Rok')
ax.set_ylabel('Populacja [mln]')

# Ustawienie etykiet osi X, aby wywietlay tylko konkretne lata
ax.set_xticks([2000, 2001, 2002, 2003])

plt.legend()
```

```python
# Filtracja danych
ohio_data = frame[(frame['state'] == 'Ohio') & (frame['year'].isin([2001, 2002]))]
nevada_data = frame[(frame['state'] == 'Nevada') & (frame['year'].isin([2001, 2002]))]

# Tworzenie wykresu
plt.plot(ohio_data['year'], ohio_data['pop'], marker='o', linestyle='-', color='b',
        label="Ohio")
plt.plot(nevada_data['year'], nevada_data['pop'], marker='o', linestyle='-', color='r',
        label="Nevada")
plt.title('Rok vs Populacja dla Ohio i Nevady')
plt.xlabel('Rok')
plt.ylabel('Populacja [mln]')

# Ustawienie etykiet osi X, aby wywietlay tylko konkretne lata
plt.xticks([2001, 2002])
plt.ylim([1.5, 3.8])
plt.legend()
```

## Zadanie 2

### Cel:
Celem tego zadania jest wczytanie danych z pliku CSV, analiza i wizualizacja tych danych na wykresach liniowych.

### Zadania:
1. Wczytaj dane z pliku `populacja_miast.csv` do DataFrame przy uyciu biblioteki pandas.
2. Stwrz wykres liniowy, ktry pokae, jak zmieniaa si populacja Warszawy i odzi na przestrzeni lat.
3. Stwrz drugi wykres liniowy, ktry pokae zmiany populacji tylko dla lat 2002, 2011 i 2022.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Wczytanie danych z CSV
df = pd.read_csv(
    "populacja_miast.csv",
    na_values=["Brak danych", "brak danych", "NA", ""]
)

# 2) Wykres liniowy: populacja Warszawy i odzi w czasie
lodz_vis = df["d"].interpolate(method="linear")  # wypenia 2015
plt.figure(figsize=(8, 5))
plt.plot(df["Rok"], df["Warszawa"], marker="o", label="Warszawa")
plt.plot(df["Rok"], df["d"], marker="o", label="d")
plt.xticks(df["Rok"].tolist())
plt.title("Zmiana populacji Warszawy i odzi w czasie")
plt.xlabel("Rok")
plt.ylabel("Liczba mieszkacw")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# 3) Drugi wykres: tylko lata 2002, 2011 i 2022
lata_docelowe = [2002, 2011, 2022]
df_subset = df[df["Rok"].isin(lata_docelowe)].sort_values("Rok")

plt.figure(figsize=(8, 5))
plt.plot(df_subset["Rok"], df_subset["Warszawa"], marker="o", label="Warszawa")
plt.plot(df_subset["Rok"], df_subset["d"], marker="o", label="d")
plt.title("Populacja (tylko lata 2002, 2011, 2022)")
plt.xlabel("Rok")
plt.ylabel("Liczba mieszkacw")
plt.grid(True, alpha=0.3)
plt.xticks(lata_docelowe)
plt.legend()
plt.tight_layout()
plt.show()
```
