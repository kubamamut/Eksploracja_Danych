```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import statsmodels.api as sm
house = pd.read_csv("AMES_housing_Price.csv", delimiter=",")
```

```python
house.head()
```

```python
#Zbior Treningowy
subset = ['SalePrice','GrLivArea','LotArea','FullBath','BedroomAbvGr','OverallQual']
house[subset].head()
predictors = ['GrLivArea','LotArea','FullBath','BedroomAbvGr','OverallQual']
outcome = 'SalePrice'
train = house.iloc[1:1000]     # wiersze 2–1000

X_train = train[predictors]
y_train = train[outcome]

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)

RMSE_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
R2_train = r2_score(y_train, y_train_pred)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

print("\nTRAINING SET RESULTS:")
print(f"RMSE_train: {RMSE_train:.0f}")
print(f"R²_train:   {R2_train:.4f}")

```

```python
#Zbior Testowy
subset = [
    'SalePrice','GrLivArea','LotArea','FullBath','BedroomAbvGr','OverallQual']
house[subset].head()
```

```python
predictors = ['GrLivArea','LotArea','FullBath','BedroomAbvGr','OverallQual']
outcome = 'SalePrice'
test  = house.iloc[1000:1461]  # wiersze 1001–1461

X_test  = test[predictors]
y_test  = test[outcome]

y_test_pred  = model.predict(X_test)

RMSE_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
R2_test = r2_score(y_test, y_test_pred)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

print("\nTEST SET RESULTS:")
print(f"RMSE_test:  {RMSE_test:.0f}")
print(f"R²_test:    {R2_test:.4f}")
```
